```python
def calculate_mean(x):
    m=[]
    n=[]
    for i in x:
        m.append(i[0])
        n.append(i[1])
    return np.mean(m),np.mean(n)
    
```

```python
k=0
l=0
while(k<30):
    c1 =[]
    c2=[]
    for i,j in zip(x1,x2):
        d1 =((k1[0]-i)**2+(k1[1]-j)**2)**.5
        d2 =((k2[0]-i)**2+(k2[1]-j)**2)**.5
        l=l+1
        print(d1,d2)
        if (d1<d2):
            h=[]
            h.append(i)
            h.append(j)
            c1.append(h)
        else:
            h=[]
            h.append(i)
            h.append(j)
            c2.append(h)
            
            # call the Claculate_mean
            C1mx,C1my =calculate_mean(c1)
            C2mx,C2my =calculate_mean(c2)
    if (C1mx==k1[0]) and(C1my==k1[1]):
        break
    else:
        k1[0]=C1mx
        k1[1]=C1my
        k2[0]=C2mx
        k2[1]=C2my
    k=k+1
        
```

```python
x=[0,2,4,3,1]
```

```python
# with in cluster sum of square
wcss =[]
for i in x:
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter =300, n_init=10, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(x,wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans= KMeans(n_clusters=3, init='k-means++',max_iter=300,n_init=10,random_state=0)
kmeans.fit(x)
```

```python
y_kmeans=kmeans.predict(x)
```

```python
y_kmeans
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1])
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=500,c='red')
plt.title('Cluster of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
